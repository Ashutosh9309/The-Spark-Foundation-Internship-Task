```python
x = SampleSuperstore.groupby('Segment')[['Sales', 'Profit']].sum().plot.bar(color=['Red', 'Yellow'])
plt.ylabel('Sales and Profit')
plt.title('Total Sales and Total Profit in Segment')
plt.show()
```
